**Изучение библиотеки math**

```python
import math
angle_in_radians = math.pi/8 #22.5 градуса
angle_in_radians
```

**Перевод градусов в радианы**

```python
angle_in_degrees = angle_in_radians * 180 / math.pi #Переводим pi/8 в градусы
angle_in_degrees 
```

**Перевод радиан в градусы**

```python
angle_to_radians = angle_in_degrees * math.pi / 180 #Переводим 22.5 градуса в радианы
angle_to_radians 
```

**импорт NumPy и работа с массивами**
